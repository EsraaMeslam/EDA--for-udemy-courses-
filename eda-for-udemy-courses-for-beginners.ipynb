<h1>Visualizations 📊📈

```python
sns.pairplot(df) #grid of scatter plots for each pair of numerical columns 
```

```python
df__temp=df.drop(['Free/Paid','year','course_title','level','subject','Date','Free/Paid','url'],axis=1) #drop unnumerical cols
df__temp
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df__temp.corr(), annot=True) ##visualizing relationships or patterns in a dataset
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True) #display the frequency or count of values 'Prices'.
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['num_subscribers'], color='blue',fill=True, shade=True)
plt.title('num_subscribers  Distribution (Density Plot)')
plt.ylabel('Density')
plt.xlim(-10000, 50000)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
random_sample = df.sample(n=80, random_state=42) #take a random sample with size 80
sns.barplot(x=random_sample['content_duration'],y=random_sample['price'])
plt.title('content_duration Vs price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
grouped_data = df.groupby('level')['num_subscribers'].sum().reset_index()
sns.barplot(x='level', y='num_subscribers', data=grouped_data)
plt.title("sum of num_subscribers for each level")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.groupby('Free/Paid')['num_subscribers'].sum().plot(kind='bar',color=['green','r'],edgecolor='black')
plt.title('Total Subscribers for Free or Paid')
plt.show()
```
